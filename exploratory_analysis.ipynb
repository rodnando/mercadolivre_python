```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Lendo o df com os dados do mercado livre

df = pd.read_csv('data/mercado_livre.csv')

df.head(2)
```

```python
# Quantidade de linhas e colunas da base

print('Quantidade de linhas da base:', df.shape[0])
print('Quantidade de colunas da base:', df.shape[1])
```

```python
# Exibindo informações gerais sobre a base
print(df.info())
```

```python
# Porcentagem de valores faltantes por coluna
missing = df.isnull().mean() * 100
print(missing[missing > 0])
```

```python
# Calculando a média de preço por marca
grouped_data = df.groupby('brand')['price'].mean()
print(grouped_data)

# Monta o gráfico de barras
grouped_data.plot(kind='bar', figsize=(10, 5))
plt.title('Média de Preço por Marca')
plt.xlabel('Categoria')
plt.ylabel('Média')
plt.show()
```

```python
# Contagem de produtos por memória RAM
category_counts = df['ram_memory'].value_counts()
print(category_counts)

# Monta o gráfico de barras
plt.figure(figsize=(10, 5))
category_counts.plot(kind='bar', figsize=(10, 5))
plt.title('Distribuição por memória RAM')
plt.ylabel('')
plt.show()
```

```python
# Contagem de produtos com Bluetooth
category_counts = df['with_bluetooth'].value_counts()
print(category_counts)

# Monta o gráfico de barras
plt.figure(figsize=(10, 5))
category_counts.plot(kind='bar', figsize=(10, 5))
plt.title('Distribuição por disponibilização de bluetooth')
plt.ylabel('')
plt.show()
```
